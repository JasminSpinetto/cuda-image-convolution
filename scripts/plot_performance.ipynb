```python
import os
import matplotlib.pyplot as plt

os.makedirs("results", exist_ok=True)

configs = [
    "CPU",
    "10B-64T",
    "40B-64T",
    "40B-128T",
    "80B-64T",
    "1280B-32T",
    "1280B-128T",
    "2560B-32T",
    "2560B-64T"
]

times = [56.5853, 1.71076, 0.375344, 0.193686, 0.19369,
         0.101917, 0.111084, 0.110814, 0.114457]

plt.figure(figsize=(10, 5))
plt.plot(configs, times, marker='o')
plt.yscale('log')
plt.xticks(rotation=45)
plt.ylabel("Execution Time (s)")
plt.title("CPU vs GPU Convolution Performance")
plt.tight_layout()
plt.savefig("../results/performance.png", dpi=300)
plt.show()
```

```python
tile_widths = [100, 200, 500, 1000]
tile_times = [0.242224, 0.154832, 0.162993, 0.167455]

plt.figure(figsize=(8, 5))
plt.plot(tile_widths, tile_times, marker='o')
plt.xlabel("Tile Width")
plt.ylabel("Execution Time (s)")
plt.title("Tiling Performance")
plt.tight_layout()
plt.savefig("../results/tiling_performance.png", dpi=300)
plt.show()
```
